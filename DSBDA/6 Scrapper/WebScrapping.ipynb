```python
request1 = requests.get("https://www.flipkart.com/oppo-reno8t-5g-midnight-black-128-gb/p/itm9ec7deac6401c?pid=MOBGHZ9HN27URAZH&lid=LSTMOBGHZ9HN27URAZHSHPISS&marketplace=FLIPKART&store=tyy%2F4io&srno=b_1_1&otracker=hp_omu_Top%2BOffers_2_4.dealCard.OMU_DC1FYKJXNYT6_3&otracker1=hp_omu_PINNED_neo%2Fmerchandising_Top%2BOffers_NA_dealCard_cc_2_NA_view-all_3&fm=neo%2Fmerchandising&iid=042f9f0b-ebef-4b3f-910e-12975299c873.MOBGHZ9HN27URAZH.SEARCH&ppt=hp&ppn=homepage&ssid=n1v8klk39s0000001684592779626")
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup 
```

```python
#fetching reviews and comments
reviews = soup.findAll('div',{'class':'t-ZTKy'});
for review in reviews:
  print(review.get_text()+'\n')
```

```python
# overall rating
ratings = soup.find('div',{'class':'_2d4LTz'}).get_text();
print(ratings)
```

```python
#individual rating 
individual_rating = soup.findAll('div',{'class':'_3LWZlK _1BLPMq'});
for indi_rating in individual_rating:
  print(indi_rating.get_text()+'\n')
```

```python
customer_name = soup.findAll('p',{'class':'_2sc7ZR _2V5EHH'});
for cust_name in customer_name:
  print(cust_name.get_text()+'\n')
```
